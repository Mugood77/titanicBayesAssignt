```python
import pandas as pd
Location = 'titanic3.xls'
df = pd.read_excel(Location)
df.head()
```

```python
pd.crosstab(df.pclass, df.survived, margins = True)
```

```python
#probability of first class given survived = p_c1_sv = 200/500
#probability of second class given survived = p_c2_sv = 119/500
#probability of third class given survived = p_c3_sv = 181/500
```

```python
pd.crosstab(df.sex, df.survived, margins = True)
```

```python
#probability of female given survived = p_F_sv = 339/500
#probability of male given survived = p_M_sv = 161/500
```

```python
p_sv = 500/1309
p_c1 = 323/1309
p_c2 = 277/1309
p_c3 = 709/1309
p_M = 843/1309
p_F = 466/1309
p_boy = 557/1046
p_c1_sv = 200/500
p_c2_sv = 119/500
p_c3_sv = 181/500
p_F_sv = 339/500
p_M_sv = 161/500
p_boy_sv = 103/427 
```

```python
def computeBayes(s, c, survival):
    return s*c*survival
p_sv_F_c3 = computeBayes(p_F_sv, p_c3_sv, p_sv)
p_sv_M_c1 = computeBayes(p_M_sv, p_c1_sv, p_sv)
p_sv_boy_c2 = computeBayes(p_boy_sv, p_c2_sv, p_sv)
print(p_sv_F_c3)
print(p_sv_M_c1)
print(p_sv_boy_c2)
```

```python
# Therefore a Female in third class was more likely to survive than a Male in first class who was more likely to survive than a boy in second class.
```
